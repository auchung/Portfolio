### **Finding the Healthiest Restaurant: Health Score Explanation**

The health score in this analysis is a calculated metric that evaluates the nutritional balance of menu items across fast food chains. It is designed to reward higher protein content while penalizing excess calories, fat, and sodium, which are typically associated with less healthy options.

#### **Our Formula:**
$$
\text{Health Score} = \frac{\text{Protein (g)}}{\text{Calories} + \text{Total Fat (g)} + \text{Sodium (mg)}}
$$
- Higher protein increases the health score.
- Higher calories, fat, and sodium lower the health score.

#### **Interpretation:**
- A higher health score suggests a more nutritionally balanced meal, with higher protein relative to excess calories, fat, and sodium.
- A lower health score indicates meals that are calorie-dense and nutrient-poor, often containing excessive fat and sodium with minimal protein.

#### **Limitations:**
- This score does not consider other nutrients like fiber, vitamins, or sugar content.
- Some high-calorie, high-protein meals (e.g., large burgers) may score better than low-calorie meals with moderate protein, even if they are not necessarily healthier.

This metric provides a relative ranking of restaurants based on their menu compositions, helping identify chains that offer higher protein-to-calorie balance.


```python
df["health_score"] = df["protein"] / (df["calories"] + df["total_fat"] + df["sodium"])

restaurant_health_scores = df.groupby("restaurant")["health_score"].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=restaurant_health_scores.index, y=restaurant_health_scores.values, palette="rocket", hue = restaurant_health_scores.index)
plt.xticks(rotation=45)
plt.title("Average Health Score by Fast Food Chain")
plt.ylabel("Health Score")
plt.xlabel("Restaurant")
plt.show()

healthiest_restaurant = restaurant_health_scores.idxmax()
print(f"The healthiest restaurant based on this score is: {healthiest_restaurant}, with a Health Score of: {restaurant_health_scores[healthiest_restaurant]}")
```

**Average Health Score by Fast Food Chain**

**Distributions:** Chick-fil-A has the highest average health score, followed by McDonald's and Subway, while Taco Bell has the lowest.

**Outliers:** No visible extreme values, but there is a noticeable gap between the highest and lowest health scores among restaurants.

**Relationships:** Restaurants with higher health scores likely offer higher protein relative to calories, fat, and sodium, while those with lower scores may have more calorie-dense, lower-protein menu options.


# Statistical Testing

# Rationale for Conducting an ANOVA Test

We decided to conduct a one-way ANOVA test to determine whether there are statistically significant differences in health scores across fast food restaurants. The rationale behind this decision is as follows:

1. **Comparing Multiple Groups**  
   We needed a statistical method to compare health scores across multiple restaurants simultaneously. ANOVA allows us to test whether at least one restaurant differs significantly from the others without performing multiple t-tests, which could increase the risk of false positives.

2. **Evaluating Overall Variability**  
   By using ANOVA, we assessed whether the observed differences in health scores are likely due to real differences between restaurants rather than random variation in the data.

3. **Providing a Foundation for Further Analysis**  
   If the ANOVA test detects significant differences, we can then conduct pairwise comparisons, such as t-tests, to determine which specific restaurants differ from Subway in terms of health scores.

4. **Ensuring a Data-Driven Approach**  
   Rather than relying on assumptions or marketing claims, we used ANOVA to objectively evaluate whether Subway is statistically different from other fast food chains in terms of healthiness.

By performing this test, we established whether there are meaningful differences among restaurants, allowing us to conduct further analysis with confidence.

```python
from scipy.stats import f_oneway, ttest_ind
restaurants = df["restaurant"].unique()
health_scores_by_restaurant = [df[df["restaurant"] == r]["health_score"].dropna() for r in restaurants]

anova_result = f_oneway(*health_scores_by_restaurant)

subway_scores = df[df["restaurant"] == "Subway"]["health_score"].dropna()
other_scores = df[df["restaurant"] != "Subway"]["health_score"].dropna()
t_test_result = ttest_ind(subway_scores, other_scores, equal_var=False)

anova_result.pvalue, t_test_result.pvalue
```

## Statistical Test Results for Subway's Health Score

## ANOVA Test (p = 3.83e-28)
- We found that this extremely low p-value (< 0.001) indicates a statistically significant difference in health scores across restaurants.
- This confirms that at least one restaurant differs significantly from the others.

## T-Test Comparing Subway to Other Restaurants (p = 0.0003)
- We observed that the p-value (0.0003) is much smaller than 0.05, meaning there is a statistically significant difference between Subway’s health score and the combined scores of all other restaurants.
- This suggests that Subway’s menu is significantly different from other fast food chains based on this metric.

## Conclusion:
- We confirmed that there are clear differences between restaurants in terms of health scores (ANOVA result).
- We found that Subway’s health score is significantly different from the average fast food restaurant (T-test result).
- We decided that further analysis is needed to determine whether this difference suggests a healthier menu or another trend in the data.

## Rationale for Conducting a Pairwise T-Test

We decided to conduct a pairwise t-test to compare Subway's health score against other fast food chains. The rationale behind this decision is as follows:

1. **Identifying Significant Differences**  
   A one-way ANOVA confirmed that there were statistically significant differences in health scores across restaurants. However, ANOVA does not tell us which specific restaurants differ from Subway. A pairwise t-test allows us to compare Subway against each individual restaurant.

2. **Direct Hypothesis Testing**  
   Our hypothesis suggests that Subway is the healthiest option among fast food chains. By using pairwise t-tests, we can directly test whether Subway's health score is statistically different from each competitor, rather than making assumptions based on averages.

3. **Controlling for Variability**  
   Each fast food chain has a range of menu items with varying health scores. A pairwise t-test helps determine if the differences we observe are consistent and significant or if they might be due to chance.

4. **Data-Driven Comparison**  
   Instead of relying on marketing claims (e.g., "Eat Fresh"), we use statistical methods to quantify health differences in a meaningful way, ensuring our conclusions are objective and data-driven.

By conducting these pairwise t-tests, we can determine whether Subway is truly a statistically healthier option or if its health score is comparable to—or even worse than—other fast food chains.


```python
from scipy.stats import ttest_ind

subway_scores = df[df["restaurant"] == "Subway"]["health_score"].dropna()

pairwise_results = {}
for restaurant in df["restaurant"].unique():
    if restaurant != "Subway":
        other_scores = df[df["restaurant"] == restaurant]["health_score"].dropna()
        t_stat, p_val = ttest_ind(subway_scores, other_scores, equal_var=False)
        pairwise_results[restaurant] = p_val

pairwise_df = pd.DataFrame(pairwise_results.items(), columns=["Restaurant", "p-value"]).sort_values("p-value")
pairwise_df
```

# Interpretation of Pairwise T-Test Results

The p-values from the pairwise t-tests indicate whether Subway's health score is significantly different from each specific restaurant. We are using the conventional threshold for significance: **p < 0.05**.

- This ensures our conclusions are based on statistical patterns rather than marketing claims.

By using clustering, we gain a more detailed understanding of how different fast food menu items compare nutritionally, allowing us to evaluate Subway’s menu within the broader fast food industry.

```python
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

clustering_features = ["calories", "total_fat", "sodium", "protein", "health_score"]
df_clustering = df[clustering_features]

imputer = SimpleImputer(strategy="mean")
df_imputed = imputer.fit_transform(df_clustering)

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_clustering)

df_scaled = scaler.fit_transform(df_imputed)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)

df["cluster"] = kmeans.fit_predict(df_scaled)

plt.figure(figsize=(10, 6))
sns.countplot(x=df["cluster"], palette="coolwarm")
plt.xlabel("Cluster")
plt.ylabel("Number of Menu Items")
plt.title("Distribution of Fast Food Items Across Clusters")
plt.show()

cluster_summary = df.groupby("cluster")[clustering_features].mean()
cluster_summary
```

# Ethics & Privacy

The datasets that we intended on using for this analysis is publicly available and sourced from corporate-reported nutrition facts from major fast food chains. Although this ensures transparency, there exists limitations that could potentially raise concerns in our analysis. For instance, fast food companies may selectively report or round values in ways that make their menu items appear healthier. Additionally, the data does not account for how individuals modify their meals such as adding extra toppings or skipping condiments which can significantly impact the nutrition values. Since the dataset does not track who consumes what, it does not allow for an equitable health impact analysis across different populations.

Regarding potential biases in our dataset, because the dataset includes only major U.S. fast food chains, it excludes smaller and local chains that may offer different health-conscious options. The dataset may also overrepresent certain menu items that chains want to highlight, potentially skewing perceptions of their overall healthiness. Raw nutritional values per item do not account for portion sizes, which may lead to misleading comparisons. For example, a small sandwich could appear healthier than a large salad simply due to size differences. 

In order to detect these biases, before analysis, we will review missing data points, check terms of use, and ensure all restaurants are equally represented. During analysis, we will normalize and standardize nutritional values per serving size or per meal to prevent portion-size distortions and cross-check with external nutritional guidelines. After analysis, when communicating results, we will clearly convey the limitations, avoid oversimplified conclusions, and present multiple rankings based on varied health priorities.

Furthermore, regarding other issues, in order to avoid unintended consequences, we plan to included external health guidelines to contextualize our findings and to frame the results transparently and consider multiple perspectives on health so that the analysis can offer more responsible and informative insights rather than oversimplified conclusions. By presenting multiple perspectives rather than a singular “healthiest” designation, we will emphasize that nutritional needs vary and that this analysis does not account for affordability, accessibility, or individual dietary restrictions and ensure that our results are not misinterpreted as universally applicable. Addressing these ethical concerns ensures that the research remains fair, inclusive, and mindful of its impact on public perception.

### Answering the Research Question
- Subway does offer many lower-calorie, lower-fat options, aligning with its "Eat Fresh" branding.

### Final Conclusion
While Subway provides a variety of low-calorie meal options, it is not definitively the healthiest fast food chain.
